## Background Subtraction

```python
cap = cv2.VideoCapture('../videos/vtest.avi')  # Open video.

fgbg = cv2.createBackgroundSubtractorMOG2()
count = 0
length = 0

def plot_image(count, frame, mask):
    plt.subplot(4, 4, count * 2 + 1)
    plt.xticks([]); plt.yticks([])
    plt.imshow(frame)

    plt.subplot(4, 4, count * 2 + 2)
    plt.xticks([]); plt.yticks([])
    plt.imshow(mask, cmap='gray')
```

```python
plt.figure(figsize=(20, 40))

while True:
    success, frame = cap.read()  # Read a frame.
    if not success or count >= 2:
        break
        
    mask = fgbg.apply(frame)  # Apply background subtraction algorithm.
    length += 1

    if length >= 100:
        plot_image(count, frame, mask)

        length = 0
        count += 1

cap.release()
```
