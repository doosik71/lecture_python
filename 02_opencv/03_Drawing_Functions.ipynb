# Drawing Functions

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('../images/cat.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
rows, cols, channels = image.shape
plt.imshow(image)
print(rows, cols)
```

## Line

```python
_ = plt.imshow(cv2.line(image.copy(), (0,0), (cols,rows), (255,0,0), 3))
```

## Rectangle

```python
_ = plt.imshow(cv2.rectangle(image.copy(), (30,30), (cols-30,rows-30), (0,255,0), 3))
```

## Circle

```python
_ = plt.imshow(cv2.circle(image.copy(), (90,90), 60, (0,0,255), 3))
```

## Text

```python
font = cv2.FONT_HERSHEY_COMPLEX
_ = plt.imshow(cv2.putText(image.copy(), 'OpenCV',
                           (int(cols/2) - 70, int(rows/2)),
                           font, 1, (255,0,255), 2, cv2.LINE_AA))
```
