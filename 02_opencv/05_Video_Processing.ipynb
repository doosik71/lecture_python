# Video Processing

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

## Basics

```python
cap = cv2.VideoCapture('../videos/vtest.avi')

success, frame = cap.read()

if success:
    plt.imshow(frame)

cap.release()
```

## Background Subtraction

```python
plt.figure(figsize=(20, 40))

cap = cv2.VideoCapture('../videos/vtest.avi')

fgbg = cv2.createBackgroundSubtractorMOG2()
count = 0
length = 0

while True:
    success, frame = cap.read()
    if not success or count >= 6:
        break
        
    fgmask = fgbg.apply(frame)
    length += 1
    
    if length >= 100:
        plt.subplot(10, 4, count * 2 + 1)
        plt.xticks([])
        plt.yticks([])
        plt.imshow(frame)
        plt.subplot(10, 4, count * 2 + 2)
        plt.xticks([])
        plt.yticks([])
        plt.imshow(fgmask, cmap='gray')
        length = 0
        count += 1

cap.release()
```
