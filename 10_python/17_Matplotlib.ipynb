# Matplotlib

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## Plotting

```python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

print('x =', x[:5], '...')
print('y =', y_sin[:5], '...')
print('(x, y) =', list(zip(x, y_sin))[:5], '...')
```

```python
plt.plot(x, y_sin)
plt.show()
```

```python
plt.plot(x, y_cos)
plt.show()
```

```python
plt.plot(x, y_sin)
plt.plot(x, y_cos)

plt.xlabel('x')
plt.ylabel('y')
plt.title('sin(x) and cos(x)')
plt.legend(['sin(x)', 'cos(x)'])

plt.show()
```

## Subplots

```python
plt.subplot(2, 1, 1)
plt.plot(x, y_sin, 'r-')
plt.title('sin(x)')

plt.subplot(2, 1, 2)
plt.plot(x, y_cos, 'b-')
plt.title('cos(x)')

plt.subplots_adjust(hspace = 0.5)
plt.savefig('../temp/sin_cos.png')

plt.show()
```

## Scatter Plot

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

## 3D Plot

```python
from mpl_toolkits import mplot3d

x, y = np.meshgrid(np.linspace(-3, 3, 61), np.linspace(-3, 3, 61))
z = 1 / (1 + np.exp(x**2+y**2))
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, cmap='jet')
ax.view_init(50, -45)
plt.show()

print(x.shape, y.shape, z.shape)
```
