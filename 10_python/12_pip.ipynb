# pip

## About
pip is a de facto standard package-management system used to install and manage software packages written in Python.

## Searching for Packages
```bash
pip search keyword
```

````python
!pip search tensorflow
````

## Installing Packages
```bash
pip install SomePackage  # To install package system-wide. This requires administrator's privilege.
                         
pip install --user SomePackage  # To install package on user environment. This does not requires admin privilege.

pip install --user SomePackage==version  # to specify version.
```

## Uninstalling Packages
```bash
pip uninstall SomePackage
```

## Listing Packages
```bash
pip list
```

````python
!pip list
````

## Using Requirements Files
"Requirements files" are files containing a list of items to be installed using pip install.
```bash
pip freeze > requirements.txt
pip install -r requirements.txt
```

````python
!pip freeze
````
